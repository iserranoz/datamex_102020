#### Use seaborn to show the distribution of the column `Age`

```python
# your code here
bins = [10,20,50]
color = ['r','b','g']
for i in range(3):
    
    ax = sns.distplot(numerical['Age'], bins = bins[i], color = color[i], kde = True)
    ax.axvline(x = numerical['Age'].mean(), c = 'k', linestyle = '--', label = 'media')
    print('BINS: ',bins[i])
    ax.legend()
    plt.show();
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. Choose the appropriate plot to show below.

```python
# your code here
fig, ax = plt.subplots()
ax.hist(titanic['Age'], bins = 30)
ax.set_xlabel('Age')
ax.set_ylabel('Freq')
plt.axvline(np.mean(titanic['Age']), c = 'r', linestyle = '--', label = 'meedia')
plt.axvline(np.mean(titanic['Age']) + np.std(titanic['Age']), c = 'k', linestyle = '--', label = '1-desviacion')
plt.axvline(np.mean(titanic['Age']) - np.std(titanic['Age']), c = 'k', linestyle = '--', label = '1-desviacion')
plt.axvline(np.mean(titanic['Age']) + 2*np.std(titanic['Age']), c = 'k', linestyle = '-', label = '2-desviacion')
plt.axvline(np.mean(titanic['Age']) - 2*np.std(titanic['Age']), c = 'k', linestyle = '-', label = '2-desviacion')
ax.legend()
```
