````python
import pandas as pd
````

````python
df = pd.DataFrame(eart, 
                  columns = ['Fecha', 'Time', 'Latitud1',
                                  'Latitud2', 'Longitud1', 'Longitud2',
                                  'Depth- Mag','pais1', 'pais2','final','borrar2','borrar3',
                            'borrar4','borrar5'])
````

````python
#df.head(20)
````

````python
df['Fecha'] = df['Fecha'].replace(r'earthquake','', regex = True)
````

````python
df['Time'] = df['Time'].replace(r'min|hr','', regex = True)
````

````python
df.reset_index()
````

````python
df['Latitud1'] = df['Latitud1'].replace(r'\d+min|ago','', regex = True)
````

````python
#df.head(20)
````

````python
df['Latitud2'] = df['Latitud2'].replace(r'ago','', regex = True)
````

````python
#df.head(20)
````

````python
df['Latitud2'] = df['Latitud2'].replace(r'N|S','', regex = True)
````

````python
df['Latitud1'] = df['Latitud1'] + df['Latitud2']
````

````python
#df.head(20)
````

````python
df['Longitud1'] = df['Longitud1'] + df['Longitud2']
````

````python
df['Longitud1'] = df['Longitud1'].replace(r'N|S|W|E','', regex = True)
````

````python
df.pais1 = df.pais1 +df.pais2 + df.final
````

````python
#df.pais1.head(20)
````

````python
df['pais1'] = df.pais1.replace(r'[0-9]|-','', regex = True)
````

````python
#df['pais1']
````

````python
#df.columns
````

````python
df.drop(['Latitud2','Longitud2','Depth- Mag','pais2','final','borrar2', 'borrar3',
       'borrar4', 'borrar5'], axis = 1, inplace = True)
````
